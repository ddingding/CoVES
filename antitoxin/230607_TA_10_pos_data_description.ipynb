```python
frac_alive = {}
thresh = 0.5
for ham in range(11):
    df_ham = df_10pos_rc3_no_stop.loc[df_10pos_rc3_no_stop['hamming'] == ham]
    frac_alive[ham] = np.sum(df_ham.mean_lrr_norm > thresh)/len(df_ham)

plt.figure(figsize=(2.5,1.75))
ax= plt.gca()
ax.set_yscale('log')
ax.scatter(frac_alive.keys(), frac_alive.values(), color='black')
plt.xticks(range(11))
plt.grid(alpha=0.3)
plt.ylabel('fraction alive (GR>0.5)')
plt.xlabel('# mutations')
plt.savefig(dout + 'frac_alive_across_mut_d_renorm.svg', format='svg')

plt.show()
print(frac_alive)
```

```python
# estimate total number of functional sequences

total_est_alive = {}
total_n_alive = 0
for n_mut in frac_alive.keys():
    
    n_possible = math.comb(10,n_mut) * (20**n_mut)
    num_muts_alive_at_thresh = n_possible * frac_alive[n_mut]
    total_est_alive[n_mut] = num_muts_alive_at_thresh
    total_n_alive +=num_muts_alive_at_thresh

print(total_n_alive/1e10, 'x1e10 number seqs total')
plt.figure(figsize=(2.5,1.75))
ax= plt.gca()
ax.scatter(total_est_alive.keys(), total_est_alive.values(), color='black')
ax.set_yscale('log')
plt.xticks(range(11))
plt.grid(alpha=0.3)
plt.ylabel('est. total # alive (GR>0.5)')
plt.xlabel('# mutations')
plt.savefig(dout + 'total_alive_across_mut_d_renorm.svg', format='svg')

plt.show()
```

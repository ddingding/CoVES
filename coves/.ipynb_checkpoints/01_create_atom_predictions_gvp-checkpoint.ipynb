# creating amino acid preference predictions from structure
- take PDB file and convert to LMDB dataset (following https://github.com/drorlab/atom3d/tree/master/atom3d)
- then take the trained RES-GVP model to find atom predictions

This code is adapted from the Atom3d and the GVP paper:

    - ATOM3D:
    - https://arxiv.org/pdf/2012.04035.pdf
    - https://github.com/drorlab/atom3d
    
    - GVP:
    - https://arxiv.org/abs/2106.03843, https://openreview.net/forum?id=1YLJDvSx6J4
    - https://github.com/drorlab/gvp-pytorch

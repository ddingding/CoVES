# 3 positions antitoxin library

```python
x,y, df_at_3p = dt.get_x_y_at_3p()
df_train, df_test = rt.get_df_train_test(df_at_3p, train_frac=0.8)
```

```python
df_train, df_test, model_subset = rt.fit_train_test(
    df_train,
    df_test,
    fit_val='E3',
    oh_col_name='oh_code',
    xlim=None,
    ylim=None,
    lr=0.1,
    epochs=300,
    batch_size=1000,
    layer=rt.log_layer(1),
    chpt_id=str(
         "at_3p_"+ str(int(time.time()))
    ),  # to make unique checkpoint
)
```

```python
rt.plot_corr_marginal(df_test.E3, df_test.yhat_non, s=0.01)
```

```python
# calculate ROC
df_yhats_y = pd.DataFrame({'yobs':df_test.E3, 
                           'yhat_non':df_test.yhat_non, 
                        })
rt.plot_roc(df_yhats_y, 
            fname = dout + 'ROC_3p',
           n_thresh=100,
           true_thresh=0.5)
```

```python
# calculate positive predictive value

rt.get_ppv(df_yhats_y, threshold=0.5)
```

# correlate logistic regression preferences and structure-based scores

```python
x,y, df_3p_at = dt.get_x_y_at_3p()
```

```python
model_3p, history, callbacks = rt.fit_log_model(x, y, 
                                        my_layer=rt.log_layer(1),
                                        adam_lr =0.1, 
                                         epochs=200,
                                         batch_size=1000)

Yhat_3pos_non= model_3p.predict(x).flatten()
```

```python
dic_pos_to_wtaa_pos = {61:'D61', 64:'K64', 80:'E80'}
dic_i_to_mut_p = {1:61, 2:64, 3: 80}
p_to_mut_to_w_lr_at_3p = rt.get_logreg_w(model_3p,rt.AA_ALPHABET, dic_i_to_mut_p)
```

```python
df_gvp_pred_poi =  df_gvp_pred.loc[df_gvp_pred.pos.isin(list(dic_i_to_mut_p.values()))]  
df_gvp_pred_poi['lr_w'] = df_gvp_pred_poi.apply(lambda r:p_to_mut_to_w_lr_at_3p[int(r.pos)][r.mut_aa], axis=1)
```

```python
score_col_n_plot = 'mean_x'
ct.plot_log_reg_vs_coves(
    df_gvp_pred_poi, 
    dic_i_to_mut_p, 
    dic_pos_to_wtaa_pos,
    fout = dout + f'3pos_lr_w_vs_{score_col_n_plot}.svg',
    n_rows=1
)
```
